```python
df.columns
```

```python
ax = sns.countplot(x= 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gn = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gn
```

```python
sg = sns.barplot(x='Gender', y= 'Amount', data=sales_gn )
```

```python
From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than the males 
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

```python
sns.barplot(x= 'Age Group', y='Amount', data = sales_age)
```

```python
From above graphs we can see that most of the buyers are of age group between 26-35 yrs females
```

```python
#total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state.head(10)
```

```python
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(x='State', y='Orders', data = sales_state)
```

```python
#total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

```python
From above graphs we can see that most of the orders are from Uttar Pradesh, Maharastra and Karnataka respectively
```
